```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your unemployment dataset into a Pandas DataFrame
unemployment_data = pd.read_csv('/content/Unemployment in India.csv')


```

```python
mean_unemployment_rate = unemployment_data[' Estimated Unemployment Rate (%)'].mean()
median_unemployment_rate = unemployment_data[' Estimated Unemployment Rate (%)'].median()
```

```python
# Plot data
plt.figure(figsize=(10, 6))
unemployment_data[' Date'] = pd.to_datetime(unemployment_data[' Date'])

# Extract month and year
unemployment_data['Month'] = unemployment_data[' Date'].dt.strftime('%b %Y')
unemployment_data['Month'] = unemployment_data['Month'].astype(str)
plt.plot(unemployment_data['Month'], unemployment_data[' Estimated Unemployment Rate (%)'], marker='o', linestyle='-')
plt.xlabel('Month')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Estimated Unemployment Rate Over Time')
plt.xticks(rotation=50)  # Rotate x-axis labels for better visibility
plt.grid()
plt.show()
```

```python
unemployment_data['Month'] = unemployment_data[' Date'].dt.strftime('%b %Y')
unemployment_data['Month'] = unemployment_data['Month'].astype(str)
plt.plot(unemployment_data['Month'], unemployment_data[' Estimated Labour Participation Rate (%)'], marker='o', linestyle='-')
plt.xlabel('Month')
plt.ylabel(' Estimated Labour Participation Rate (%)')
plt.title(' Estimated Labour Participation Rate (%) Over Time')
plt.xticks(rotation=100)  # Rotate x-axis labels for better visibility
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(unemployment_data[' Estimated Unemployment Rate (%)'], bins=20)
plt.xlabel('Estimated Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.title('Distribution of Estimated Unemployment Rate')
plt.grid()
plt.show()
```

```python
# Plot data
plt.figure(figsize=(10, 6))

unemployment_data['Area'] = unemployment_data['Area'].astype(str)
plt.plot(unemployment_data['Area'], unemployment_data[' Estimated Unemployment Rate (%)'], marker='o', linestyle='-')
plt.xlabel('Area')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Estimated Unemployment Rate Over Area')
plt.xticks(rotation=50)  # Rotate x-axis labels for better visibility
plt.grid()
plt.show()
```
